```python
#Scaling
def Scale(data,feild):
    tenure = data[feild]
    tenure_np = np.array(tenure).reshape(-1,1)
    scaler = MinMaxScaler()
    tenure_scale = scaler.fit_transform(tenure_np)
    tenure_scaled = pd.DataFrame(tenure_scale)
    
    return tenure_scaled
```

## Model creation 

```python
 X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.2)
```

```python
 # 1587 is the MAXIMUM
```

```python
# Scale(data,feild):
```

```python
def first_check(model,X):
    check = model.predict(X)
    print(check.sum())
```

```python
#Apply Random forest 
rf_model = RandomForestClassifier()
rf_model.fit(X_train,y_train)

y_train_rf_pred = rf_model.predict_proba(X_train)# keep probabilities for the positive outcome only
print('RF train roc-auc: {}'.format(roc_auc_score(y_train, y_train_rf_pred[:,1]))) #This is the compares y_train to the second column i.e. where it redicted 1
#So if I do  ytrain_pred[:,0], the output is (1-ytrain_pred[:,1] )

ytest_rf_pred = rf_model.predict_proba(X_test)
print('RF test roc-auc: {}'.format(roc_auc_score(y_test, ytest_rf_pred[:,1])))
```

```python
first_check(rf_model,X)
```

```python
#xgb_model
xgb_model = xgb.XGBClassifier(learning_rate=0.1,max_depth=6,n_estimators=500,n_jobs=-1)
xgb_model.fit(X_train, y_train)

ytrain_pred_xg = xgb_model.predict_proba(X_train)
print('XG train roc-auc: {}'.format(roc_auc_score(y_train, ytrain_pred_xg[:,1]))) 

ytest_pred_xg = xgb_model.predict_proba(X_test)
print('XG test roc-auc: {}'.format(roc_auc_score(y_test, ytest_pred_xg[:,1])))
```

```python
first_check(xgb_model,X)
```

```python
#LogisticRegression
log_classifier=LogisticRegression()
log_classifier.fit(X_train, y_train)

ytrain_pred_log = log_classifier.predict_proba(X_train)
print('Logistic Regression train roc-auc: {}'.format(roc_auc_score(y_train, ytrain_pred_log[:,1]))) 

ytest_pred_log = log_classifier.predict_proba(X_test)
print('Logistic Regression test roc-auc: {}'.format(roc_auc_score(y_test, ytest_pred_log[:,1]))) 
```

```python
first_check(log_classifier,X)
```

```python
#KNeighborsClassifier
knn_classifier=KNeighborsClassifier()
knn_classifier.fit(X_train, y_train)

ytrain_pred_knn = knn_classifier.predict_proba(X_train)
print('KNN train roc-auc: {}'.format(roc_auc_score(y_train, ytrain_pred_knn[:,1]))) 

ytest_pred_knn = knn_classifier.predict_proba(X_test)
print('KNN test roc-auc: {}'.format(roc_auc_score(y_test, ytest_pred_knn[:,1])))
```

```python
first_check(knn_classifier,X)
```

```python
#AdaBoostClassifier
ada_classifier=AdaBoostClassifier()
ada_classifier.fit(X_train, y_train)

ytrain_pred_ada = ada_classifier.predict_proba(X_train)
print('Adaboost train roc-auc: {}'.format(roc_auc_score(y_train, ytrain_pred_ada[:,1])))

ytest_pred_ada = ada_classifier.predict_proba(X_test)
print('Adaboost test roc-auc: {}'.format(roc_auc_score(y_test, ytest_pred_ada[:,1])))
```

**Accuracy** - Accuracy is the most intuitive performance measure and it is simply a *ratio of correctly predicted observation to the total observations.*

**Precision** - Precision is the ratio of *correctly predicted positive observations to the total predicted positive observations.*  Of all passengers that labeled as survived, how many actually survived? This check if something was labeled 1, was it actally a 1. **Of all the 1s the model predicted, how many of these were actually 1.**
 

Precision = TP/TP+FP

**High precision relates to the low false positive rate** Makes sure the model does not give false positives 


Accuracy - Accuracy is the most intuitive performance measure and it is simply a ratio of correctly predicted observation to the total observations.

**Recall(Sensitivity)** -  Recall is the ratio of *correctly predicted positive observations to the all observations* in actual class.  Of all the passengers that truly survived, how many did we label? **Of actual 1s, how many was the model able to predict.**


Recall = TP/TP+FN

Aviodes false negatives 

To make sure the models dont miss potential customers, I will be using recall is the important thing 

```python
grid_sc = GridSearchCV(estimator =  RandomForestClassifier() ,
                   param_grid = {'class_weight': [{0:1 , 1:m} for m in np.linspace(1,20,15)]}, #between 1 and 20 with 15 gaps inbetween 
                   scoring = {'precision': make_scorer(precision_score), 'recall': make_scorer(recall_score)}, #This adds botth to the grid search 
                   refit = 'recall' ,#The model should focus on precision     
                   cv = 8 , 
                   n_jobs =1)
```

```python
random_2 = grid_sc.fit(X,y)
```

```python
pd.DataFrame(random_2.cv_results_)
```
